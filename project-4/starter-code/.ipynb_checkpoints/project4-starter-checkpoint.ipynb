### How do you deal with the prestige feature?

I would want to make dummy variables out of the prestige feature. Because there are four different prestige categories, I would want to create three different dummy variables to avoid issues of multicollinearity. 

### What is the best model and that model's parameter's for predicting admission? Hint: Use GridsearchCV and define what metric you use to find what is best.

```python
from sklearn.linear_model import LogisticRegression
```

```python
lm = LogisticRegression()
data_raw = data_raw.dropna()
dummy_ranks = pd.get_dummies(data_raw['prestige'],prefix="prestige")
cols_to_keep = ['admit', 'gre', 'gpa']
data_raw = data_raw[cols_to_keep].join(dummy_ranks.ix[:, 'prestige_1':])
data = data_raw[cols_to_keep].join(dummy_ranks.ix[:, 'prestige_2':])

print data.head()

data['intercept'] = 1.0 #need to manually insert intercept column
```

```python
print data_raw.head()
```

```python
train_cols = data.columns[1:]
y = data["admit"]
```

```python
from sklearn import grid_search, cross_validation
from sklearn.metrics import roc_curve, roc_auc_score

c_space = np.logspace(-5, 8, 15)
param_grid = {'C': c_space}

gs = grid_search.GridSearchCV(
    estimator = LogisticRegression(), 
    param_grid = param_grid,
    scoring = 'roc_auc',
    cv = 5)
gs.fit(data[train_cols], y)
```

```python
print gs.best_score_ 
```

```python
print gs.best_estimator_.coef_
```

```python
print gs.best_estimator_
```

### What is the most important feature that contributes to the students probability of admission? 

The coefficient with the greatest absolute value is that belonging to prestige = 4 (or belonging to the school with the lowest ranking). The take-away from this is that going to a low-ranking school greatly reduces your odds of graduate school admission. 

### Visualize the model relative to prestige and GPA. HINT: Use the technique from the visualization of the flight delay model

```python
data_raw['probability'] = gs.predict_proba(data[train_cols]).T[1]
```

```python
features = ['prestige_1.0', 'prestige_2.0', 'prestige_3.0', 'prestige_4.0', 'gpa']
features
```

```python
import numpy as np
import matplotlib.pyplot as plt

ax1 = plt.subplot(111)
colors = ['blue', 'yellow', 'red', 'purple']
for e, color in enumerate(colors):
    data_raw[data_raw[features[e]] == 1].plot(x='gpa', y='probability', kind='scatter', c=color, ax=ax1, figsize=(15,15))
ax1.set(title='Probability of Admission\n Based on Previous School Ranking and GPA')

lines, labels = ax1.get_legend_handles_labels()
plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20)
font = {'family' : 'normal',
       'weight' : 'bold',
       'size'   : 22}

plt.rc('font', **font)
# ax1.suptitle('test title', fontsize=20)
ax1.legend(lines[:2], labels=features, loc='upper left',fontsize=15)  # legend for first two lines only
```
